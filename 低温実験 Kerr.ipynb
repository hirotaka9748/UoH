半波長板を回転させる

```python
import room138.module138.λ
import room138.module138.pico
import room138.module138.KIKUSUI
import os
import time
import datetime

num  = "1"

deg_start = 300   #モーター移動開始（度）
deg_stop  = 480  #モーター移動終了（度）
deg_step  = 5    #刻み（度）
detect_time = 10      #積算回数

################################################################################

t_delta = datetime.timedelta(hours=9)
JST = datetime.timezone(t_delta, 'JST')
now = datetime.datetime.now(JST)
date = now.strftime('%Y_%m_%d')

status = 'H:/マイドライブ/Data/' + date + '/status_log/'  #モーター位置
pico1  = 'H:/マイドライブ/Data/' + date + '/pico1_log/'   #ピコアン１

status_path  = status + num + '.txt' 
pico_path1   =  pico1 + num + '.txt' 

mean   = 'H:/マイドライブ/Data/' + date + '/mean/'
result = 'H:/マイドライブ/Data/' + date + '/result/'
calc   = 'H:/マイドライブ/Data/' + date + '/calc/'

pico_num_path1 = pico1 + num + '_num' + '.txt' 

status_csv_path = status + num + '.csv' 
pico_csv_path1  =  pico1 + num + '.csv' 
mean_path1     =   mean + num + '_path1' + '.csv'
result_path    = result + num + '.csv'
calc_path      =   calc + num + '.csv'

os.makedirs(status, exist_ok=True)
os.makedirs(pico1,  exist_ok=True)
os.makedirs(mean,  exist_ok=True)
os.makedirs(result,exist_ok=True)
os.makedirs(calc,  exist_ok=True)

range_start = float(deg_start)/0.0025
range_stop  = float(deg_stop)/0.0025
range_step  = float(deg_step)/0.0025

room138.module138.KIKUSUI.teslaOFF()
room138.module138.KIKUSUI.currON()

room138.module138.λ.move0(int(range_start))

for i in range(int(range_start), int(range_stop), int(range_step)):
    room138.module138.λ.move(i)
    room138.module138.λ.status(status_path)
    i = 0
    while i < detect_time:
        room138.module138.pico.pico1(pico_path1)
        i += 1

room138.module138.pico.numpico(pico_path1,pico_num_path1)
room138.module138.calc.toCSV(pico_num_path1,pico_csv_path1)
room138.module138.calc.mean(detect_time,pico_csv_path1,pico_num_path1,mean_path1)
room138.module138.calc.motorStatus(status_path,status_csv_path)
room138.module138.calc.copyStatus(status_csv_path,calc)
room138.module138.calc.copyPico1(mean_path1,calc)
room138.module138.calc.link(calc,calc_path)
room138.module138.calc.removeStatus(calc)
room138.module138.calc.removePico1(calc)
%matplotlib
room138.module138.plot.pandasPlots(calc_path)
room138.module138.calc.move(calc_path,result)
```

```python
import room138.module138.KIKUSUI
import room138.module138.pico
import room138.module138.calc
import room138.module138.plot
import os
import datetime

num  = "2"

detect_time = 20   #積算回数

##################################################################################

t_delta = datetime.timedelta(hours=9)
JST = datetime.timezone(t_delta, 'JST')
now = datetime.datetime.now(JST)
date = now.strftime('%Y_%m_%d')

pico1  = 'H:/マイドライブ/Data/' + date + '/pico1_log/'   #ピコアン１
pico2  = 'H:/マイドライブ/Data/' + date + '/pico2_log/'   #ピコアン２
mean   = 'H:/マイドライブ/Data/' + date + '/mean/'
atan   = 'H:/マイドライブ/Data/' + date + '/atan/'
diff   = 'H:/マイドライブ/Data/' + date + '/diff/'
result = 'H:/マイドライブ/Data/' + date + '/result/'
calc   = 'H:/マイドライブ/Data/' + date + '/calc/'
kerr   = 'H:/マイドライブ/Data/' + date + '/kerr/'

pico_path1 = pico1 + num + '.txt' 
pico_path2 = pico2 + num + '.txt' 

pico_num_path1 = pico1 + num + '_num' + '.txt' 
pico_num_path2 = pico2 + num + '_num' + '.txt' 

pico_csv_path1  =  pico1 + num + '.csv' 
pico_csv_path2  =  pico2 + num + '.csv' 
mean_path1     =   mean + num + '_path1' + '.csv'
mean_path2     =   mean + num + '_path2' + '.csv'
atan_path      =   atan + num + '.csv'
diff_path      =   diff + num + '.csv'
result_path    = result + num + '.csv'
calc_path      =   calc + num + '.csv'
kerr_path      =   kerr + num + '.csv'

os.makedirs(pico1, exist_ok=True)
os.makedirs(pico2, exist_ok=True)
os.makedirs(mean,  exist_ok=True)
os.makedirs(atan,  exist_ok=True)
os.makedirs(diff,  exist_ok=True)
os.makedirs(result,exist_ok=True)
os.makedirs(calc,  exist_ok=True)
os.makedirs(kerr,  exist_ok=True)

i = 0
while i < detect_time:
    room138.module138.pico.pico1(pico_path1)
    room138.module138.pico.pico2(pico_path2)
    i += 1

room138.module138.pico.numpico(pico_path1,pico_num_path1)
room138.module138.pico.numpico(pico_path2,pico_num_path2)
room138.module138.calc.toCSV(pico_num_path1,pico_csv_path1)
room138.module138.calc.toCSV(pico_num_path2,pico_csv_path2)
room138.module138.calc.mean(detect_time,pico_csv_path1,pico_num_path1,mean_path1)
room138.module138.calc.mean(detect_time,pico_csv_path2,pico_num_path2,mean_path2)
room138.module138.calc.arctan(mean_path1,mean_path2,atan_path)
room138.module138.calc.diff(atan_path,diff_path)
room138.module138.calc.kerr(diff_path,kerr_path)
```

磁石を動かしながらKerr回転角測定

```python
import room138.module138.KIKUSUI
import room138.module138.pico
import room138.module138.calc
import room138.module138.plot
import room138.module138.teion
import os
import datetime

num  = "5"
step  = 1 #mm
detect_time = 50   #積算回数

##################################################################################

t_delta = datetime.timedelta(hours=9)
JST = datetime.timezone(t_delta, 'JST')
now = datetime.datetime.now(JST)
date = now.strftime('%Y_%m_%d')

pico1  = 'G:/マイドライブ/ユーザー/安部/' + date + '/pico1_log/'   #ピコアン１
pico2  = 'G:/マイドライブ/ユーザー/安部/' + date + '/pico2_log/'   #ピコアン２
mean   = 'G:/マイドライブ/ユーザー/安部/' + date + '/mean/'
atan   = 'G:/マイドライブ/ユーザー/安部/' + date + '/atan/'
result = 'G:/マイドライブ/ユーザー/安部/' + date + '/result/'
calc   = 'G:/マイドライブ/ユーザー/安部/' + date + '/calc/'
kerr   = 'G:/マイドライブ/ユーザー/安部/' + date + '/kerr/'


motor1 = 'G:/マイドライブ/ユーザー/安部/' + date + '/motor1/'
motor2 = 'G:/マイドライブ/ユーザー/安部/' + date + '/motor2/'
motor = 'G:/マイドライブ/ユーザー/安部/' + date + '/motor/'

pico_path1 = pico1 + num + '.txt' 
pico_path2 = pico2 + num + '.txt' 

pico_num_path1 = pico1 + num + '_num' + '.txt' 
pico_num_path2 = pico2 + num + '_num' + '.txt'
motor_path1    = motor1 + num + '.txt'
motor_path2    = motor2 + num + '.txt'
motor_path     = motor + num + '.txt'

pico_csv_path1  =  pico1 + num + '.csv' 
pico_csv_path2  =  pico2 + num + '.csv' 
mean_path1     =   mean + num + '_path1' + '.csv'
mean_path2     =   mean + num + '_path2' + '.csv'
atan_path      =   atan + num + '.csv'
result_path    = result + num + '.csv'
calc_path      =   calc + num + '.csv'
kerr_path      =   kerr + num + '.csv' 
motor_csv_path = motor + num + '.csv' 

os.makedirs(pico1, exist_ok=True)
os.makedirs(pico2, exist_ok=True)
os.makedirs(mean,  exist_ok=True)
os.makedirs(atan,  exist_ok=True)
os.makedirs(result,exist_ok=True)
os.makedirs(calc,  exist_ok=True)
os.makedirs(kerr,  exist_ok=True)
os.makedirs(motor,  exist_ok=True)
os.makedirs(motor1,exist_ok=True)
os.makedirs(motor2,exist_ok=True)

range_start = 0
range_stop  = 25000
range_step  = float(step)*1000


for i in range(range_start, range_stop+1, int(range_step)):
    room138.module138.teion.move_m1(i)
    room138.module138.teion.status_m1(motor_path1)
    room138.module138.teion.status_m2(motor_path2)
    a = 0
    while a < detect_time:
        room138.module138.pico.pico1(pico_path1)
        room138.module138.pico.pico2(pico_path2)
        a += 1
for j in range(range_start+int(range_step), range_stop+1, int(range_step)):
    room138.module138.teion.move_m2(j)
    room138.module138.teion.status_m1(motor_path1)
    room138.module138.teion.status_m2(motor_path2)
    b = 0
    while b < detect_time:
        room138.module138.pico.pico1(pico_path1)
        room138.module138.pico.pico2(pico_path2)
        b += 1
        
room138.module138.teion.move_m1(0)
room138.module138.teion.move_m2(0)
        
room138.module138.calc.sum_status(motor_path1,motor_path2,motor_path)
room138.module138.pico.numpico(pico_path1,pico_num_path1)
room138.module138.pico.numpico(pico_path2,pico_num_path2)
room138.module138.calc.toCSV(motor_path,motor_csv_path)
room138.module138.calc.toCSV(pico_num_path1,pico_csv_path1)
room138.module138.calc.toCSV(pico_num_path2,pico_csv_path2)
room138.module138.calc.mean(detect_time,pico_csv_path1,pico_num_path1,mean_path1)
room138.module138.calc.mean(detect_time,pico_csv_path2,pico_num_path2,mean_path2)
room138.module138.calc.arctan(mean_path1,mean_path2,atan_path)
room138.module138.calc.diffkerr(atan_path,kerr_path)
room138.module138.calc.copyTesla(motor_csv_path,calc)
room138.module138.calc.copyDiff(kerr_path,calc)
room138.module138.calc.link(calc,calc_path)
room138.module138.calc.removeTesla(calc)
room138.module138.calc.removeDiff(calc)
%matplotlib
room138.module138.plot.pandasPlot(calc_path)
room138.module138.calc.move(calc_path,result)

```
