温度測定

```python
import socket
import time
import datetime
import os

num = "1"
path = 'C:/Users/wadat/Data1/2021_12_21/temp_log/'
path1 = path + num + ".txt"

os.makedirs(path, exist_ok=True)
i = 0

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    # サーバを指定
    s.connect(('150.12.70.76', 7020))
    
    while i < 100:
        dt_now = datetime.datetime.now()
        print(dt_now)
        with open(path1, 'a') as f:
            print(dt_now, file=f)
        
        # サーバにメッセージを送る
        s.sendall(b"READ:DEV:MB1.T1:TEMP\n")
        # ネットワークのバッファサイズは1024。サーバからの文字列を取得する
        data = s.recv(1024)
        Meas1 = repr(data)
        Meas2 = repr(data).strip('b')
        Meas3 = Meas2.replace("'","")
        Meas4 = Meas3[452:480]
        print(Meas4)
        time.sleep(15)
        i += 1
        with open(path1, 'a') as f:
            print(Meas4, file=f)
```
