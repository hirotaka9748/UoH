```python
def f(x, y, mu, S):
  x_norm = (np.array([x, y]) - mu[:, None, None]).transpose(1, 2, 0)
  return np.exp(- x_norm[:, :, None, :] @ np.linalg.inv(S)[None, None, :, :] @ x_norm[:, :, :, None] / 2.0) / (2*np.pi*np.sqrt(np.linalg.det(S)))
```

```python
# condig: utf-8
import os, sys
import numpy as np
import cv2
import matplotlib
from matplotlib import pylab as plt
from mpl_toolkits.mplot3d import Axes3D

class ImageColorProcessing(object):

   def __init__(self, image_path=None):
        """  """
       self.image_path = image_path
       self.image = self.read_image(self.image_path)
   
   def read_image(self, image_path):
       """  """
       image = cv2.imread(image_path)
       if image is None:
           print("Image[{}] can not open.".format(image_path))
           sys.exit(1)
       return image
   
   def bgr_to_rgb_to_flat(self):
       """ BGR -> BGB -> 3ch"""
       rgb_image = cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB) # BGR -> RGB
       height, width, channel = rgb_image.shape
       rgb_flat = rgb_image.reshape((height*width, channel)) # 
       return rgb_flat

   def duplication_rgb(self):
       """ RGB """
       flat = self.bgr_to_rgb_to_flat()
       duplication, counts = np.unique(flat, return_counts=True, axis=0)
       return duplication, counts.reshape((counts.shape[0], 1))
   
   def plot3d(self, x, y, z, size, color):
       """ 3 """
       fig = plt.figure()
       ax = fig.add_subplot(111, projection="3d")
       ax.scatter(x, y, z, s=size, c=color)
       ax.set_xlim([0, 255])
       ax.set_ylim([0, 255])
       ax.set_zlim([0, 255])
       ax.set_xlabel("R")
       ax.set_ylabel("G")
       ax.set_zlabel("B")
       plt.show()

   def rgb_plot3d(self):
       """ RGB """
       duplication, counts = self.duplication_rgb()
       red, green, blue = np.hsplit(duplication, [1, 2])
       self.plot3d(red, green, blue, counts, duplication/255.)
   
   def rgbData_to_csv(self):
       """ RGBCSV """
       duplication, counts = self.duplication_rgb()
       rgbData = np.hstack([duplication, counts])
       np.savetxt("rgb.csv", rgbData, delimiter=",", header="r,g,b,counts")

def main():

    ICP = ImageColorProcessing(image_path="***") # ***
    #ICP.rgb_plot3d() # 
    #ICP.rgbData_to_csv() # 

if __name__ == "__main__": main()
```

```python
import sys
import numpy as np
import cv2
from matplotlib import pyplot as plt


# 
drawing = False
complete_region = False
ix,iy,width,height = -1,-1,0,0
box = [ix, iy, width, height]


# 
def my_mouse_callback(event,x,y,flags,param):
    global ix,iy,width,height,box,drawing,complete_region

    if event == cv2.EVENT_MOUSEMOVE:      # 
        if(drawing == True):
            width = x - ix
            height = y - iy

    elif event == cv2.EVENT_LBUTTONDOWN:  # 
        drawing = True

        ix = x
        iy = y
        width = 0
        height = 0

    elif event == cv2.EVENT_LBUTTONUP:    # 
        drawing = False
        complete_region = True

        if(width < 0):
            ix += width
            width *= -1
        if(height < 0):
           iy += height
           height *= -1

    box = [ix, iy, width, height]         # 




# 
def main():
    global ix,iy,width,height,box,drawing,complete_region

    source_window = "draw_rectangle"
    roi_window = "region_of_image"

    img = cv2.imread(sys.argv[1],1)  # 
    temp = img.copy()                # 

    cv2.namedWindow(source_window)
    cv2.setMouseCallback(source_window, my_mouse_callback)

    while(1):
        cv2.imshow(source_window,temp)

        if(drawing):             # 
            temp = img.copy()    # 
            cv2.rectangle(temp,(ix,iy),(ix + width, iy+ height),(0,255,0),2)  # 

        if(complete_region): # 
            complete_region = False

            roi = img[iy:iy+height, ix:ix+width] # 
            cv2.imshow(roi_window, roi)          # 

            # 
            color = ('b','g','r')
            for i,col in enumerate(color):
                histr = cv2.calcHist([roi],[i],None,[256],[0,256])
                plt.plot(histr,color = col)
                plt.xlim([0,256])
            plt.show(block=False)

        # 
        k = cv2.waitKey(1) & 0xFF
        if k == 27:          # esc
            break
        elif k ==ord('s'):   # 's'
            cv2.imwrite('roi.png', roi)
            cv2.imwrite('draw_src.png', temp)
            plt.savefig('histgram.png')

    cv2.destroyAllWindows()



if __name__ == "__main__":
    main()
```

```python
##
# cording:utf-8
##

import numpy as np
import cv2

def main():
    # 
    img = cv2.imread("test.jpg")

    print(img.shape)

    # (left(x1), top(y1), right(x2), bottom(y2))
    roi = (937,218, 1280, 350)

    # ()
    dst_img = img.copy()

    # ROI
    # [top:bottom, left:right] 
    s_roi = img[roi[1]: roi[3], roi[0]: roi[2]]
    s_roi = cv2.blur(s_roi, (30, 30)) # 

    #
    dst_img[roi[1]: roi[3], roi[0]: roi[2]] = s_roi

    # 
    rect_img = img.copy()
    cv2.rectangle(rect_img, (roi[2],roi[3]), (roi[0],roi[1]), (0, 255, 0), 2)

    # 
    cv2.namedWindow("image", cv2.WINDOW_AUTOSIZE | cv2.WINDOW_FREERATIO)
    cv2.namedWindow("dst_image", cv2.WINDOW_AUTOSIZE | cv2.WINDOW_FREERATIO)
    cv2.imshow("image", rect_img)
    cv2.imshow("dst_image", dst_img)
    cv2.waitKey()

    cv2.destroyAllWindows()

if __name__ == '__main__':
    main()
```
